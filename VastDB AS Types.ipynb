```python
import csv
import numpy as np
from collections import defaultdict 
from operator import itemgetter 
```

```python
with open('Rebuttal/VAST_DB-hg19') as bed, \
    open('Rebuttal/Ex_up.bed', "w") as ex_up, \
    open('Rebuttal/Ex_down.bed', "w") as ex_down, \
    open('Rebuttal/Ex_up_LC.bed', "w") as ex_up_LC, \
    open('Rebuttal/Ex_down_LC.bed', "w") as ex_down_LC, \
    open('Rebuttal/Ex_up_IC.bed', "w") as ex_up_IC,  \
    open('Rebuttal/Ex_down_IC.bed', "w") as ex_down_IC, \
    open('Rebuttal/Ex_up_HC.bed', "w") as ex_up_HC, \
    open('Rebuttal/Ex_down_HC.bed', "w") as ex_down_HC, \
    open('Rebuttal/ALTD_1.bed', "w") as ALTD_1, \
    open('Rebuttal/ALTD_2.bed', "w") as ALTD_2, \
    open('Rebuttal/ALTA_1.bed', "w") as ALTA_1, \
    open('Rebuttal/ALTA_2.bed', "w") as ALTA_2, \
    open('Rebuttal/MULTI1_up.bed', "w") as MULTI1_up, \
    open('Rebuttal/MULTI1_down.bed', "w") as MULTI1_down, \
    open('Rebuttal/MULTI2_up.bed', "w") as MULTI2_up, \
    open('Rebuttal/MULTI2_down.bed', "w") as MULTI2_down :


    reader = csv.reader(bed, delimiter='\t')
     
    for row in reader:

        csv.field_size_limit(1000000000)
        transcript = row[3]
        
        ID = row[3]

        qstarts = list(map (int, row[11].strip(",").split(",")))                      
        blocksizes = list(map(int, row[10].strip(",").split(",")))
        start = int(row[1])
        strand = row[5]
        bn = int(row[9])
        chr = row[0]
        
        score = row[8]
        
        intermediate_complexity = "170,0,255"
        high_complexity = "255,0,0"
        low_complexity = "0,0,255"
        
        
        if "EX" in ID:
            
            q1 = qstarts[1]
            b = blocksizes[1]
            
            estart = start + q1
            eend = start + q1 + b

            Ex_up = [chr, estart, estart+1,  ID, 0, strand]
            Ex_down = [chr, eend, eend+1,  ID, 0, strand]
            
            if strand=="-":
                Ex_up =  [chr, eend, eend+1, ID, 0, strand]
                Ex_down = [chr, estart, estart+1, ID, 0, strand]
                
            ex_up.write("\t".join(map(str, Ex_up)) + "\n")
            ex_down.write("\t".join(map(str, Ex_down)) + "\n")
            
            if score==low_complexity:
                ex_up_LC.write("\t".join(map(str, Ex_up)) + "\n")
                ex_down_LC.write("\t".join(map(str, Ex_down)) + "\n")
                
            if score==intermediate_complexity:
                ex_up_IC.write("\t".join(map(str, Ex_up)) + "\n")
                ex_down_IC.write("\t".join(map(str, Ex_down)) + "\n")
    
            if score==high_complexity:
                ex_up_HC.write("\t".join(map(str, Ex_up)) + "\n")
                ex_down_HC.write("\t".join(map(str, Ex_down)) + "\n")
            
        
        if "ALTD" in ID:
            
            if strand == "+":
            
                n = int(ID.split("-")[-1].split("/")[0])

                q1 = qstarts[0]
                b = blocksizes[0]
                estart = start + q1
                eend = start + q1 + b

                donor = [chr, eend, eend+1,  ID, 0, strand]


                if n==1:
                    ALTD_1.write("\t".join(map(str, donor)) + "\n")
                if n==2:
                    ALTD_2.write("\t".join(map(str, donor)) + "\n")
            
            
            if strand == "-":
            
                n = int(ID.split("-")[-1].split("/")[0])

                q1 = qstarts[-1]
                b = blocksizes[-1]
                estart = start + q1
                eend = start + q1 + b

                donor = [chr, estart, estart+1,  ID, 0, strand]

                if n==1:
                    ALTD_1.write("\t".join(map(str, donor)) + "\n")
                if n==2:
                    ALTD_2.write("\t".join(map(str, donor)) + "\n")


        if "ALTA" in ID:
        
            if strand == "+":
            
                n = int(ID.split("-")[-1].split("/")[0])

                q1 = qstarts[-1]
                b = blocksizes[-1]
                estart = start + q1
                eend = start + q1 + b

                acceptor = [chr, estart, estart+1,  ID, 0, strand]


                if n==1:
                    ALTA_1.write("\t".join(map(str, acceptor)) + "\n")
                if n==2:
                    ALTA_2.write("\t".join(map(str, acceptor)) + "\n")
            
            
            if strand == "-":
            
                n = int(ID.split("-")[-1].split("/")[0])

                q1 = qstarts[0]
                b = blocksizes[0]
                estart = start + q1
                eend = start + q1 + b

                acceptor = [chr, eend, eend+1,  ID, 0, strand]

                if n==1:
                    ALTA_1.write("\t".join(map(str, acceptor)) + "\n")
                if n==2:
                    ALTA_2.write("\t".join(map(str, acceptor)) + "\n")
                                
        if ID=="MULTI(2)":
            
            i = 0
            
            for q1, b in zip(qstarts[1:-1], blocksizes[1:-1]):
                
                i+=1
                
                estart = start + q1
                eend = start + q1 + b
                
                exon = chr + ":" +  str(estart) + strand + str(eend)
                
                if strand =="+":
                
                    acceptor = [chr, estart, estart+1,  ID, 0, strand]
                    donor = [chr, eend, eend+1,  ID, 0, strand]
                
                if strand == "-":
                    acceptor = [chr, eend, eend+1,  ID, 0, strand]
                    donor = [chr, estart, estart+1,  ID, 0, strand]
                
                if i==1:
                    MULTI1_up.write("\t".join(map(str, acceptor)) + "\n")
                    MULTI1_down.write("\t".join(map(str, donor)) + "\n")
                if i==2:
                    MULTI2_up.write("\t".join(map(str, acceptor)) + "\n")
                    MULTI2_down.write("\t".join(map(str, donor)) + "\n")                
 

                
    
```

# Alternative exons

```python
qual_samples = dict()
COORDs = dict([])

with open('Rebuttal/PSI_TABLE-hg19.tab') as PSI_file:
    
    reader = csv.DictReader(PSI_file, delimiter="\t")
    
    headers = reader.fieldnames
    
    
    samples = []
    
    
    for col in headers[6:]:
    
        if "-Q" in col:
            
            sample = col.split("-")[0]
            qual_samples[sample] = col
            
        else:
            samples.append(col)
            
            
    event_tissues = defaultdict(list)
            
    
    for row in reader:
        
        ID = row["EVENT"]
        COORDs[row["COORD"]] = ID
    
        for s in samples:
            
            qual = row[qual_samples[s]].split(",")[0]
            
            tissue = s.split("_")[0]
            
            
            
            if qual=="SOK" or qual=="OK":
                
                PSI = float(row[s])
                
                event_tissues[(tissue, ID)].append(PSI)
  
```

```python
event_tissue_mean = defaultdict(list)

for key in event_tissues:
    
    tissue, ID = key
    mean_PSI = np.mean(event_tissues[key])
    
    event_tissue_mean[ID].append(mean_PSI)
```

```python
EX_max_MADs = []
INT_max_MADs = []
ALTA_max_MADs = []
ALTD_max_MADs = []

for ID in event_tissue_mean:
    
    PSIs = event_tissue_mean[ID]
    
    if len(PSIs)>=5:
        
        median = np.median(PSIs)
        
        MADs = [ abs(x-median) for x in PSIs]
        
        if "EX" in ID:
            EX_max_MADs.append( (ID, max(MADs)) )
        elif "INT" in ID:
            INT_max_MADs.append( (ID, max(MADs)) )
        elif "ALTA" in ID:
            ALTA_max_MADs.append( (ID, max(MADs)) )            
        elif "ALTD" in ID:
            ALTD_max_MADs.append( (ID, max(MADs)) )             
            
```

```python
EX_max_MADs.sort(key=lambda x:x[1])
INT_max_MADs.sort(key=lambda x:x[1])
ALTA_max_MADs.sort(key=lambda x:x[1])
ALTD_max_MADs.sort(key=lambda x:x[1])
```

```python
def split(a, n):
    k, m = divmod(len(a), n)
    return (a[i*k+min(i, m):(i+1)*k+min(i+1, m)] for i in range(n))
```
